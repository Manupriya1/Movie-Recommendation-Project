# Import Libraries

```python
import pandas as pd
```

```python
movies = pd.read_csv("movies_metadata.csv", low_memory=False)
movies.head(5)
```

```python
ratings = pd.read_csv("ratings.csv", low_memory=False)
ratings.head(5)
```

```python
movie = pd.read_csv("movies.csv", low_memory=False)
movie.head(5)
```

```python
df_merge = pd.merge(movie, ratings, how = "left", left_on = "movieId", right_on = "movieId")
df_merge.head(5)
```

```python
movies_final = pd.merge(movies, df_merge,  how = "left",  left_on = "title", right_on = "title")
print(movies_final.head(5))
movies_final.info()
```
